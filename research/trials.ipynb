```python
import os
import sys

# Force reload modules to get fresh imports
for mod in list(sys.modules.keys()):
    if 'langchain' in mod or 'pinecone' in mod:
        del sys.modules[mod]

from pinecone import Pinecone, ServerlessSpec

# Use absolute imports to avoid conflicts
import pypdf
from pathlib import Path

# Step 1: Load PDF files manually
print("Loading PDF files...")
data_path = r"F:\My Projects\End_to_End_Medical_Chatbot\data"
documents = []

for pdf_file in Path(data_path).glob("**/*.pdf"):
    print(f"  Loading {pdf_file.name}...")
    with open(pdf_file, 'rb') as f:
        reader = pypdf.PdfReader(f)
        for page_num, page in enumerate(reader.pages):
            text = page.extract_text()
            documents.append({
                "content": text,
                "source": pdf_file.name,
                "page": page_num
            })

print(f"Loaded {len(documents)} pages from PDFs")

# Step 2: Split text into chunks
print("Splitting text into chunks...")
chunk_size = 500
chunk_overlap = 20
text_chunks = []

for doc in documents:
    text = doc["content"]
    for i in range(0, len(text), chunk_size - chunk_overlap):
        chunk = text[i:i + chunk_size]
        if chunk.strip():
            text_chunks.append({
                "content": chunk,
                "source": doc["source"],
                "page": doc["page"]
            })

print(f"Created {len(text_chunks)} text chunks")

# Step 3: Download embeddings
print("Loading embedding model...")
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("all-MiniLM-L6-v2")
print("Embedding model loaded")

# Step 4: Initialize Pinecone
print("Initializing Pinecone...")
api_key = os.environ.get("PINECONE_API_KEY")
pc = Pinecone(api_key=api_key)
index_name = "textbot"

existing_indexes = pc.list_indexes().names()
if index_name not in existing_indexes:
    print(f"Creating index '{index_name}'...")
    pc.create_index(
        name=index_name,
        dimension=384,
        metric="cosine",
        spec=ServerlessSpec(
            cloud="aws",
            region="us-east-1"
        )
    )
    print(f"Index '{index_name}' created successfully")
else:
    print(f"Index '{index_name}' already exists")

# Step 5: Upload documents to Pinecone
print("Uploading documents to Pinecone...")
index = pc.Index(index_name)

batch_size = 100
for i in range(0, len(text_chunks), batch_size):
    batch = text_chunks[i:i + batch_size]
    vectors = []
    
    for j, chunk in enumerate(batch):
        embedding = model.encode(chunk["content"]).tolist()
        vector_id = f"chunk_{i + j}"
        vectors.append({
            "id": vector_id,
            "values": embedding,
            "metadata": {
                "text": chunk["content"][:300],
                "source": chunk["source"],
                "page": chunk["page"]
            }
        })
    
    index.upsert(vectors=vectors)
    print(f"  Uploaded {len(vectors)} vectors ({i + len(batch)}/{len(text_chunks)})")

print("✓ Documents successfully uploaded to Pinecone!")
```
